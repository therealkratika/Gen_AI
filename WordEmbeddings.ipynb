```python
#Implementing Dimensionally Reduction
#- we have to reduce dimension from 100D to 2D
#- because we cannot visualize see:
#Similar words cluster together
# Opposite words separate
# Semantic relationships
```

```python

#We can reduce Dimensions by using:
#1. PCA- Linear Reduction
#- PCA finds direction where data varies most
#-Directions= Principal Components

# Suppose students:
# -height and weight are correlated
# -instead of 2 features- PCA create 1 new feature representing both

# Steps of PCA:
# 1.Standardize data
# 2.Compute covariance matrix
# 3.Find eigenvectors
# 4.Select top components
# 5.Project the data

#2. t-SNE- Non Linear
# t-distributed Stochastic Neighbor Embedding
# Preserves local structure (neighbors)
# Points close in high dim -> close in low dim
# steps for SNE
#3. UMAP
```

```python
!pip install gensim
import gensim.downloader as api
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
model = api.load("glove-wiki-gigaword-100")
print("Model loaded successfully..")
```

```python
words = [
    "king","queen","man","woman",
    "dog","cat","animal",
    "car","truck","vehicle",
    "happy", "sad","joy", "anger"
]
vectors = [model[word] for word in words]
```

```python
pca = PCA(n_components=2)
reduced_vectors = pca.fit_transform(vectors)
```

```python
plt.figure(figsize=(10,10))
for i, word in enumerate(words):
  x = reduced_vectors[i][0]
  y = reduced_vectors[i][1]
  plt.scatter(x,y)
  plt.text(x + 0.01, y+ 0.01, word)
  plt.title("Word Embeddings Visualization")
plt.show()
```

```python
from sklearn.manifold import TSNE
import numpy as np
```

```python
vectors = np.array([model[word] for word in words])
```

```python
tsne = TSNE(n_components=2,perplexity=3)
reduced_vectors = tsne.fit_transform(vectors)
```

```python
plt.figure(figsize=(10,10))
for i, word in enumerate(words):
  x = reduced_vectors[i][0]
  y = reduced_vectors[i][1]
  plt.scatter(x,y)
  plt.text(x + 0.01, y+ 0.01, word)
plt.title("Word Embeddings TSNE")
plt.show()
```
